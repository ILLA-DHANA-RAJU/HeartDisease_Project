```python
data_base = data.copy()
data_base.head()
```

```python
data_base['Gender'] = data_base['Gender'].replace({ 1 : 'Men' , 0 : 'Women'})
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=data_base, x='Gender', hue='HeartDisease')
plt.title('Gender Sepearation data')
plt.show()
```

```python
Young_age = data_base[(data_base['Gender'].isin(['Men','Women'])) & (data_base['Age'].between(30,40))]
Middle_age = data_base[(data_base['Gender'].isin(['Men','Women'])) & (data_base['Age'].between(40,60))]
Old_age = data_base[(data_base['Gender'].isin(['Men','Women'])) & (data_base['Age'].between(60,78))]
```

```python
print("Younger age pepole:",len(Young_age))
```

```python
print("Middle aged pepole:",len(Middle_age))
```

```python
print("Older age pepole:",len(Old_age))
```

```python
plt.figure(figsize=(10, 4))
sns.histplot(data=data_base, x='Age', kde=True , hue='HeartDisease')
plt.title('Age Ploting Heart Disease')
plt.show()
```

```python
data_base['HeartDisease'] = data_base['HeartDisease'].replace({ 1 : 'Positive' , 0 : 'negitive'})
```

```python
plt.figure(figsize=(40, 15))
sns.countplot(data= data_base, x= 'Age', hue = "HeartDisease")
plt.title('Figure Over all Age HeartDisease Data')
plt.show()
```

```python
Men_Deisease_positive = data_base[(data_base['Gender'] == 'Men') & (data_base['HeartDisease'] == 'Positive' )]
Men_Deisease_negative = data_base[(data_base['Gender'] == 'Men') & (data_base['HeartDisease'] == 'negitive' )]
Women_Deisease_positive = data_base[(data_base['Gender'] == 'Women') & (data_base['HeartDisease'] == 'Positive' )]
Women_Deisease_negative = data_base[(data_base['Gender'] == 'Women') & (data_base['HeartDisease'] == 'negitive' )]
```

```python
print("Total Positive Men:", len(Men_Deisease_positive))
```

```python
print("Total Negative Men:", len(Men_Deisease_negative))
```

```python
print("Total Positive Women:", len(Women_Deisease_positive))
```

```python
print("Total Negative Women:", len(Women_Deisease_negative))
```

```python
plt.figure(figsize=(15, 4))
sns.histplot(data=Men_Deisease_positive , x='Age', bins=50, kde=True, color='Red')
plt.xlabel("Age")
plt.ylabel("Count of Positive Men")
plt.title("Heart Disease in Men by Age")
plt.show()
```

```python
plt.figure(figsize=(15, 4))
sns.histplot(data=Men_Deisease_negative , x='Age', bins=50, kde=True, color='blue')
plt.xlabel("Age")
plt.ylabel("Count of Negative Men")
plt.title("Heart Disease in Men by Age")
plt.show()
```

```python
plt.figure(figsize=(15, 4))
sns.histplot(data=Women_Deisease_positive , x='Age', bins=50, kde=True, color='Red')
plt.xlabel("Age")
plt.ylabel("Count of Positive Women")
plt.title("Heart Disease in Women by Age")
plt.show()
```

```python
plt.figure(figsize=(15, 4))
sns.histplot(data=Women_Deisease_negative, x='Age', bins=50, kde=True, color='blue')
plt.xlabel("Age")
plt.ylabel("Count of Negative Women")
plt.title("Heart Disease in Women by Age")
plt.show()
```

```python
plt.figure(figsize=(40, 10))
sns.countplot(data=data_base, x='HeartRate', hue='HeartDisease')
plt.title('Accurate data')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.boxplot(data=data_base, x='Gender', y='HeartRate', hue='HeartDisease')
plt.ylim(0, 200)
plt.title('Heart Rate base Dividing data')
plt.show()
```
